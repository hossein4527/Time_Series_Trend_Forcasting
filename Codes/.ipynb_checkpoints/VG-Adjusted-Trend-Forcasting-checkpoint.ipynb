```python
def most_similarity_matrix(time_series, T=20, tp=50):
    G = visibility_graph(time_series)
    adj_matrix = np.array(adjacency_matrix(G).todense())
    E = G.number_of_edges()
    k = adj_matrix.sum(1)
    P = adj_matrix/k
    n = k.shape[0]
    s_srw = np.zeros((n,n))
    pi = np.diag(np.ones(n))
    for t in range(T):
        pi = P@pi
        s_srw += (k*pi+k.T*pi.T)/(2*E)
    if tp != 0:
        triup = np.triu(s_srw,k=-tp)
        triup[triup != 0] = 1
        s_srw = np.tril(s_srw,k=-1) * triup
    return s_srw
```

```python
def error_bars( selected_data, forcast_ts):
    RMSE = math.sqrt(np.square(np.subtract(selected_data,forcast_ts)).mean())
    NRMSE = 100*RMSE/(max(selected_data)-min(selected_data))
    MAPE=np.mean(np.abs((selected_data - forcast_ts)/selected_data))*100
    MAP=np.mean(np.abs((selected_data - forcast_ts)))
    return {'RMSE':RMSE, 'NRMSE':NRMSE, 'MAPE':MAPE, 'MAP':MAP}
```

```python
def Cross_Vis_Graph(ts1, ts2):
    def is_edge_cross(ts1 , ts2 , t1,t2):
        t1 , t2 = min(t1,t2) , max(t1,t2)
        for i in range(t1+1,t2):
            if ts2[i] >= ts2[t2] + (ts1[t1]-ts1[t2])*(t2-i)/(t2-t1):
                return False
        return True

    VG = nx.Graph()
    for i in range(len(ts1)):
        print(str(i)+'/'+str(len(ts1))+'  = '+str(np.round(i/len(ts1),2)*100)+'%')
        for j in range(i+1,len(ts1)):
            if is_edge_cross(ts1,ts2, i,j):
                VG.add_edge(i,j)
    return VG.edges   

def node_similarity(Graph_networkx, T=20, tp=30):
    G = Graph_networkx
    adj_matrix = np.array(adjacency_matrix(G).todense())
    E = G.number_of_edges()
    k = adj_matrix.sum(1)
    P = adj_matrix/k
    n = k.shape[0]
    s_srw = np.zeros((n,n))
    pi = np.diag(np.ones(n))
    for t in range(T):
        pi = P@pi
        s_srw += (k*pi+k.T*pi.T)/(2*E)
    if tp != 0:
        triup = np.triu(s_srw,k=-tp)
        triup[triup != 0] = 1
        s_srw = np.tril(s_srw,k=-1) * triup
    return s_srw, np.argmax(s_srw,axis=1)
```
