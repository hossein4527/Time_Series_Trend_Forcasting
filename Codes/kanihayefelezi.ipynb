```python
import pandas as pd
```

```python
orig_df = pd.read_excel('../Data/Tehran_Indexes_Price_NoRegime.xlsx').set_index('date')
```

```python
kani_df = pd.read_excel('../Data/kani.xlsx').set_index(pd.to_datetime(kani_df['date']))
```

```python
kani_df.drop('date', axis=1, inplace=True)
```

```python
kani_df = kani_df[::-1]
```

```python
final_df = pd.merge(orig_df, kani_df, how='outer', on='date').dropna()
```

```python
final_df.to_excel('TH.xlsx')
```

```python
import numpy as np
```

```python
def price_return(p):
    pr = []
    for i in range(1, len(p)):
        pr.append(np.log(p[i])-np.log(p[i-1]))
    return pr
```

```python
complete_network_df_returned = final_df.copy()
for i in range(len(complete_network_df_returned.columns)):
    returned_col = price_return(complete_network_df_returned[complete_network_df_returned.columns[i]].values)
    returned_col.append(0)
    returned_col = np.array(returned_col)
    returned_col = returned_col - np.mean(returned_col)
    returned_col = returned_col/np.std(returned_col)
    complete_network_df_returned[complete_network_df_returned.columns[i]] = returned_col
```

```python
complete_network_df_returned.iloc[:-1].to_excel('TTH.xlsx')
```
